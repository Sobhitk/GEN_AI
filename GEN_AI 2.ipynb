2. Probabilistic Modeling
>>Generative models use probability distributions to generate new data similar to training data.
Difference:
Generative Model → Learns data distribution (P(X))
Discriminative Model → Learns decision boundary (P(Y|X))

```python
#6. VAE for Text (Simplified Code)
from tensorflow.keras.preprocessing.text import Tokenizer
from tensorflow.keras.preprocessing.sequence import pad_sequences

texts = ["The sky is blue", "The sun is bright", "The grass is green",
         "The night is dark", "The stars are shining"]

tokenizer = Tokenizer()
tokenizer.fit_on_texts(texts)
sequences = tokenizer.texts_to_sequences(texts)

padded = pad_sequences(sequences)

print(padded)
```

```python
#7. GPT Translation
from transformers import AutoTokenizer, AutoModelForSeq2SeqLM

text = "AI is transforming the world."

# French Translation
model_name_fr = "Helsinki-NLP/opus-mt-en-fr"
tokenizer_fr = AutoTokenizer.from_pretrained(model_name_fr)
model_fr = AutoModelForSeq2SeqLM.from_pretrained(model_name_fr)

inputs_fr = tokenizer_fr(text, return_tensors="pt")
outputs_fr = model_fr.generate(**inputs_fr)
translated_text_fr = tokenizer_fr.decode(outputs_fr[0], skip_special_tokens=True)
print(f"English to French: {translated_text_fr}")

# German Translation
model_name_de = "Helsinki-NLP/opus-mt-en-de"
tokenizer_de = AutoTokenizer.from_pretrained(model_name_de)
model_de = AutoModelForSeq2SeqLM.from_pretrained(model_name_de)

inputs_de = tokenizer_de(text, return_tensors="pt")
outputs_de = model_de.generate(**inputs_de)
translated_text_de = tokenizer_de.decode(outputs_de[0], skip_special_tokens=True)
print(f"English to German: {translated_text_de}")
```
